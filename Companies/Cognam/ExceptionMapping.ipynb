```python
import pandas as pd
import numpy as np
```

```python
#read as string to avoid type conversion
df = pd.read_csv("./newdata.csv", dtype=str)
```

```python
#display all rows
pd.set_option('display.max_rows', None)
#replace NaN with empty string
df = df.replace(np.nan, '', regex=True)
```

```python
#array of error code
error_code = df['ErrorCode'].unique()
```

```python
#sort error code
error_code.sort()

# remove 2nd and 3rd element and append to the end
error_code = np.append(error_code[3:], error_code[:3])

#remove 00000000 from error code and place it at beginning
error_code = np.delete(error_code, np.where(error_code == '00000000'))
error_code = np.insert(error_code, 0, '00000000')
```

```python
error_code
```

```python
# dictionary with error code as key and new error code as value
# 000000000 -> 0000 and rest in range 6001 - 6999 except
error_code_mapping = {}
for i in range(len(error_code)):
    if error_code[i] == '00000000':
        error_code_mapping[error_code[i]] = '0000'
    else:
        error_code_mapping[error_code[i]] = str(6000 + i)
```

```python
error_code_mapping
```

```python
#map error code to new error code
df['NewErrorCode'] = df['ErrorCode'].map(error_code_mapping)
```

```python
#save to csv
df.to_csv('newMap.csv', index=False)
```
