```python
import pandas as pd
```

```python
classDF = pd.read_csv("./apierrorcode.csv", dtype=str)
```

```python
#remove rows with ResponseCode value of NaN
classDF = classDF[classDF.ResponseCode.notnull()]
```

```python
#replace nan with empty string
classDF = classDF.fillna('')

#display all rows
pd.set_option('display.max_rows', None)
classDF
```

```python
#column names
classDF.columns
```

```python
# dictionar of classDF with ResponseCode as key and Exception as value
classDict = dict(zip(classDF.ResponseCode, classDF.ExceptionName))
```

```python
classDict
```

```python
df = pd.read_csv("./file1.csv", dtype=str)
```

```python
# replace nan with empty string
df = df.fillna('')
```

```python
df
```

```python
#column names
df.columns
```

```python
# add class name based on mapping in classDict
df['Class Name'] = df['Resp Code'].map(classDict) 
```
